# Data Visualization Using Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Line chart
x = [5,7,1,9,2]
y = [10,20,30,40,50]
plt.plot(x,y, color='yellow')
plt.xlabel('Line Data')
plt.ylabel('Distribution')
plt.title('My line chart')
plt.show()
```

```python
# Scatter plot
plt.scatter(x, y, color='black')
plt.xlabel('Data')
plt.ylabel('Distribution')
plt.title('My scatter')
plt.show()
```

```python

plt.bar(x,y, color='green')
plt.xlabel('Data')
plt.ylabel('Distribution')
plt.title('My Bar')
plt.show()
```

### Comparing two bar chart

```python
plt.bar([1,4,7,5,3],[4,8,3,1,4], label='First bar')
plt.bar([2,5,1,8,6],[2,4,6,8,10], label='Second bar', color='red')
plt.legend()
plt.title('Comparing two bar chart')
plt.xlabel('Data')
plt.ylabel('Distribution')
plt.show()
```

```python
plt.hist(x, color='red')
plt.xlabel('Serial NO for y data:')
plt.ylabel('Distribution')
plt.title('My histogram')
plt.show()
```

```python
# scatter with 1000 data
import numpy as np
x=np.random.random((1000))
y=np.linspace(1,100,1000)# this will generate 1000 numbers from 1 to 100
plt.scatter(x,y, color='red')
plt.title('Scatter with 1000 dots')
plt.xlabel('Data')
plt.ylabel('Distribution')
plt.show()
```

# Three Dimensional Plotting with Matplotlib (3D)

# 3D Plots

```python
from mpl_toolkits import mplot3d
```

```python
ax = plt.axes(projection='3d')
z=np.linspace(0,1,100)
x=z*np.cos(20*z)
y=z*np.sin(20*z)
ax.plot3D(x,y,z, 'green')
ax.set_title('3D line plot')
plt.show()
```

```python
ax=plt.axes(projection = '3d')
z=np.linspace(1,100,100)
x=z*np.cos(200*z)
y=z*np.sin(200*z)
ax.plot3D(x,y,z, 'red')
ax.set_title('3D line plot')
plt.show()
```

```python
ax=plt.axes(projection='3d')
z=np.linspace(1,100, 1000)
x=z*np.cos(3000*z)
y=z*np.sin(3000*z)
ax.plot3D(x,y,z, 'yellow')
ax.set_title('3D line plot')
plt.show()
```

# 3D Countour Plot

```python
ax=plt.axes(projection='3d')
x=np.linspace(-6, 6, 50)
y=np.linspace(-6, 6, 50)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2 + y**2))

ax.contour3D(x,y,z, 200, cmap='Accent')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Countour')
plt.show()
```

## 3D Surface Plot

```python
ax=plt.axes(projection = '3d')
x=np.linspace(-6, 6, 30)
y=np.linspace(-6, 6, 30)
x, y  = np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2 +y**2))

ax.plot_surface(x,y,z, color='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface')
plt.show()
```
