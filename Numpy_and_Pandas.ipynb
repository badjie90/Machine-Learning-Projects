## Indexing Using Labels in Pandas

#### Indexing can be worked with labels using "Pandas.DataFrame.loc" method, which allows to index using labels instead of positions, Using loc, we can also slice the pandas dataframe over a range of indices.

```python
# Selecting data from one row to another row
df2.loc[1:4]
```

```python
# selecting the first rows including the 5th index and every columns of our dataset
df2.loc[0:5,:]
```

```python
# Compute various statistical values
df2.describe()
```

**Select rows with indices Using "iloc"**

**When we are using iloc, we need to specify the rows and columns by their iteger index. If we want to select only the first abd third row, we simply need to put this into a list in the iloc statement with our dataframe:**

```python
# Select rows with indexs
# Here we select the 0 row and the 2nd row
df2.iloc[[0,2]]
```

```python
# Selecting the 1st and the 10th row
df2.iloc[[1,10]]
```

```python
# Select rows with a particular indexes and particular columns
df2.iloc[[0,2,4,6,8,10],# rows
         [1,3]]# columns
```
