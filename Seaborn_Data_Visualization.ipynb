# Data Visualization with Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

## Distribution Plot

```python
sn.distplot([0,1,2,3,4,5])
plt.show()
```

```python
# Generation a random numbers and plot the distribution plot
np.random.seed(0)
a = np.random.randn(200)
sn.distplot(a, color='red')
plt.show()
```

## Line Plot

```python
x=[10, 5, 30, 20, 70]
y=[20, 50,100, 60,80]
data_plot =pd.DataFrame({'Product_no':x, 'Cost':y})
sn.lineplot(x='Product_no', y='Cost', data=data_plot)
plt.show()
```

```python
x=[1,2,4,7,9,10]
y=[4,9,4,2,1,6,]
sn.lineplot(x,y)
plt.show()
```
